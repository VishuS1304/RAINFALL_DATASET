```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
nRowsRead = 1000
rf = pd.read_csv('/content/RJrainfall_data.csv', delimiter=',', nrows = nRowsRead)
rf.dataframeName = '/content/RJrainfall_data.csv'
nRow, nCol = rj.shape
print(f'There are {nRow} rows and {nCol} columns')
```

```python
rf.head(5)
```

```python
df = pd.DataFrame(rj)

DISTRICT = df['DISTRICT'].head(20)
ANNUAL = df[('ANNUAL')].head(34)

# Figure Size
fig = plt.figure(figsize =(20, 8))

# Horizontal Bar Plot
plt.bar(DISTRICT[0:10], ANNUAL[0:10])
plt.xlabel('District')
plt.ylabel('Annual')
plt.legend()
# Show Plot
plt.show()
```

```python
df = pd.DataFrame(rj)
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
JAN = df[('JAN')].head(34)
FEB = df[('FEB')].head(34)

# Set position of bar on X axis
br1 = np.arange(len(JAN))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
plt.bar(br1, JAN , color ='r', width = barWidth,
		edgecolor ='grey', label ='IT')
plt.bar(br2, FEB, color ='g', width = barWidth,
		edgecolor ='grey', label ='ECE')

# Adding Xticks
plt.xlabel('JAN/FEB', fontweight ='bold', fontsize = 15)
plt.ylabel('RAINFALL IN mm', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(JAN))],
		[''])

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,5))
rj.groupby(['DISTRICT'])['ANNUAL'].sum().sort_values(ascending=False).head(12).plot(kind='line', color = 'g')
plt.ylabel('Yearly Rainfall')
plt.title('Yearly Rainfall Data')
```

```python
plt.figure(figsize=(15,5))
rj.groupby(['DISTRICT'])['ANNUAL'].sum().sort_values(ascending=False).head(12).plot(kind='hist', color = 'b')
plt.ylabel('Yearly Rainfall')
plt.title('Yearly Rainfall Data')
```

```python
plt.figure(figsize=(15,5))
rj.groupby(['DISTRICT'])['ANNUAL'].sum().sort_values(ascending=False).head(12).plot(kind='bar', color = 'y')
plt.ylabel('Total Rainfall')
plt.title('Total Rainfall Data')
```

```python
rf.groupby("DISTRICT").sum()['ANNUAL'].plot(figsize=(12,8));
```

```python
#here we used groupby function and sorted the values in ascending order to create rainfall data
plt.figure(figsize=(15,5))
rf.groupby(['DISTRICT','ANNUAL'])['JJAS'].sum().sort_values(ascending=False).plot()
plt.xticks(rotation=90)
```
